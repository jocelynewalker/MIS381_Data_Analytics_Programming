```python
## Question 1

import random

# create empty dictionary to store human answers
human_answers = {"Rock":0,"Paper":0,"Scissors":0}

# create value to store number of games played
game_count = 0

# create dictionary of winners
winners = {"Human":0,"Computer":0, "Draw": 0}

# Create empty variable for human_answer
human_answer = ""

# define function to determine computer answers
def determine_computer_answer(human_answers):
    # checks for the maximum value of the human answers
    maxVal = max(human_answers.values())
    # pulls all the maximum values in the case there's a tie
    keys = [key for key, value in human_answers.items() if value == maxVal]
    # randomly chooses which answer to try to beat if there's multiple
    choice = random.choice(keys)
    
    #determines computer's answer in order to try to beat human's expected choice
    if choice == "Rock":
        computer_answer = "Paper"
    elif choice == "Paper":
        computer_answer = "Scissors"
    else:
        computer_answer = "Rock"
        
    return computer_answer

# define function to get human answer
def determine_human_answer(human_answers):
    answer = input("\nChoose Rock, Paper, Scissors, or Quit: ")
    
    # validates user input
    while answer != "Rock" and answer != "Paper" and answer != "Scissors" and answer != "Quit":
        print("Input not valid. Please try again.")
        answer = input("\nChoose Rock, Paper, Scissors, or Quit: ")
    
    # adds the answer choice to the dictionary tracker if it's not 'quit'
    if answer != "Quit":
        human_answers[answer] += 1
    return answer

# define function to determine winner
def check_who_wins(computer_answer, human_answer, winners):
    # checks for draw and records draw result 
    if(computer_answer == human_answer):
        print("\n Your answer:", human_answer, "\n Computer answer:", computer_answer,"\n Result is a draw.")
        winners["Draw"] += 1
        
    # based on human answer, determines winning computer answer and returns results
    else:
        if(human_answer=="Rock"):
            if(computer_answer=="Paper"):
                winners = computer_wins(human_answer, computer_answer, winners)
            else:
                winners = human_wins(human_answer, computer_answer, winners)
        elif(human_answer=="Scissors"):
            if(computer_answer=="Rock"):
                winners = computer_wins(human_answer, computer_answer, winners)
            else:
                winners = human_wins(human_answer, computer_answer, winners)
        else:
            if(computer_answer=="Scissors"):
                winners = computer_wins(human_answer, computer_answer, winners)
            else:
                winners = human_wins(human_answer, computer_answer, winners) 
    return winners

# define function for if computer wins, print this result
def computer_wins(human_answer, computer_answer, winners):
    print("\n Your answer:", human_answer, "\n Computer answer:", computer_answer,"\n Computer wins!")
    winners["Computer"] += 1
    return winners

# define function for if human wins, print this result
def human_wins(human_answer, computer_answer, winners):
    print("\n Your answer:", human_answer, "\n Computer answer:", computer_answer,"\n Human wins!")
    winners["Human"] += 1
    return winners

# define end of game print function
def end_of_game(winners, game_count):
    print("\nNumber of games played is:",game_count)
    print("Total number of times human won is:", winners["Human"])

# run while loop as long as human answer is not quit
while human_answer != "Quit":
    # get computer answer
    computer_answer = determine_computer_answer(human_answers)
    
    # gets human answer
    human_answer = determine_human_answer(human_answers)
    
    # if human didn't answer quit, then check who wins and add to the game count
    if human_answer != "Quit":
        result = check_who_wins(computer_answer, human_answer, winners)
        game_count += 1

#when human quits, end the game
end_of_game(winners, game_count)
```

```python
## Question 2

# open html file
fp = open('FloridaVoters.html', 'r')

# reads first line 
curline = fp.readline()

# reads and throws out all lines before first line of table
while 'ALACHUA' not in curline:
    curline = fp.readline()

# create empty list
lines = []

# adds first stripped line to list
lines.append(curline.lstrip("<th><d>").rstrip("</th>\n</td"))

# reads all other lines
for curline in fp:
    # if it gets to the last line of the table, read and throw out the rest of the lines
    if '12,215,216' in curline:
        fp.read()
    
    # if the lines are not "tr" or "r," read them
    if ('/tr' not in curline) and ('r' not in curline):
        #strip the lines of html characters and commas
        curline = curline.lstrip("<th><d>").rstrip("</th>\n</td").replace(',', '')
        # if the line is a number, convert to int
        if curline.isdigit():
            curline = int(curline)
        # add the line to the list
        lines.append(curline)

# create list that groups each line into tuples
new_lines = [tuple(lines[i:i+3]) for i in range(0, len(lines), 6)]

# define function for sort key of democrat voters
def getDemocratVoters(list_item):
    return(list_item[2])

# sorts the list
new_lines.sort(key=getDemocratVoters)

# prints the list 
for i in range(len(new_lines)):
    print(new_lines[i][0],new_lines[i][1], new_lines[i][2])

# closes file
fp.close()
```

```python
## Question 3
import math

## part a
def moving_window(tweet):
    # create empty dict
    tweet_dict = {}
    # splits tweet based on spaces
    tweet = tweet.split(" ")
    # for each word in each tweet, create a dictionary entry for its three-word phrase
    for i in range(len(tweet)-2):
        tweet_sect = ' '.join(list(tweet[i:i+3]))
        tweet_dict.update({tweet_sect:1})
    # returns the dictionary
    return tweet_dict

## part b
def cosine(dict_1, dict_2):
    # counts the number of phrases in each tweet
    n1 = len(dict_1)
    n2 = len(dict_2)
    
    #create 0 variable for match_count
    match_count = 0
    
    # for each key, check if other keys in dictionary match. if so, add to match count
    for key in dict_1.keys():
        if key in dict_2.keys():
            match_count += 1
    # calculate cosine similarity and return
    cos_similarity = match_count/math.sqrt(n1*n2)
    return cos_similarity

## part c

# read file
fp = open('Santa.txt', 'r')
# create empty list for tweets
tweets = []
# for each line, add to the tweets list
for line in fp:
    tweets.append(line.rstrip())
fp.close()

# for each tweet, check all previously seen tweets
for i in range(len(tweets)):
    for k in range(i):
        # calcate cosine similarity between this tweet and previous tweet
        cos_similarity = cosine(moving_window(tweets[i]),moving_window(tweets[k]))
        # if the cosine similarity is > 0.5, then print
        if cos_similarity > 0.5:
            print(tweets[i],"\nis near duplicate with tweet\n",tweets[k], "\n")
```

```python
## Question 4
import re
import math

# Part a - build a list of full quotes
fp = open("quotes.txt", "r", encoding= 'utf-8', errors='ignore')
quotes = []
for line in fp:
    quotes.append(line.rstrip() + " - " + next(fp).rstrip())
fp.close()

# part b - words from full quotes
def quote_words(quote):
    words = re.split('\W+',quote.lower().replace('`','').replace('-','').replace('$','').replace(',',''))
    return words

# part c - build the postings-list dictionary
postings_dict = {}
for quote in quotes:
    value_dict = {}
    word_list = quote_words(quote)
    for word in word_list:
        value_dict.update({word:word_list.count(word)})
    postings_dict.update({quote:value_dict})

# part d - build the reverse postings-list dictionary
reverse_postings_dict = {}

for quote in quotes:
    word_list = quote_words(quote)
    for word in word_list:
        if word not in reverse_postings_dict:
            reverse_postings_dict.update({word:{quote:word_list.count(word)}})
        else:
            reverse_postings_dict[word].update({quote:word_list.count(word)})

# part e - write a tf-idf function
def tf_idf(word, quote):
        if (' ' + word + ' ') in quote:
            w_count = postings_dict[quote][word.lower()]
        else:
            w_count = 0
        max_count = max(postings_dict[quote].values())
        tf = w_count/max_count
        n = len(quotes)
        q_count = len((reverse_postings_dict[word.lower()].values()))
        idf = math.log(n/q_count)
        tf_idf = tf*idf
        return tf_idf

# part f - quote search using a single word
def single_quote_search(word):
    search_dict = {}
    quote_list = list(reverse_postings_dict[word].keys())
    for quote in quote_list:
        value = tf_idf(word, quote)
        search_dict.update({quote:value})
    return search_dict

single_quote_search("how")

# part g - quote search using multiple words
def multiple_quote_search(word_list):
    search_dict = {}
    quote_list = []
    for word in word_list:
        inner_list = list(reverse_postings_dict[word].keys())
        for quote in inner_list:
            quote_list.append(quote)
    
    for quote in quote_list:
        value = 0
        for word in word_list:
            if (' ' + word + ' ') in quote:
                value += tf_idf(word, quote)
        search_dict.update({quote:value})
    return search_dict

multiple_quote_search(["heart", "mind"])
```
